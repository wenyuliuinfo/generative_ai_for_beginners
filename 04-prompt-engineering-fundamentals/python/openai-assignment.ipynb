# Prompt Engineering
Prompt Engineering is the process of designing and optimizing prompts for natural language processing tasks. It involves selecting the right prompts, tuning their parameters, and evaluating their performance.


### Exercise 1: Tokenization
Explore Tokenization using tiktoken, an open-source fast tokenizer from OpenAI.

````python
# Import library
import tiktoken

# Define the prompt you want to tokenized
text = f"""
Generative AI is artificial intelligence capable of generating text, \
images and other types of content. \
What makes it a fantastic technology is that it democratizes AI, \
anyone can use it with as little as a text prompt, \
a sentence written in a natural language. \
There's no need for you to learn a language, \
state what you want and out comes a suggestion from an AI model.
"""

# Set the model you want to encoding for 
encoding = tiktoken.encoding_for_model("gpt-4o-mini")

tokens = encoding.encode(text)
print(tokens);

[encoding.decode_single_token_bytes(token) for token in tokens]
````

### Exercise 2: Validate OpenAI API Key Setup
Run the code below to verify that your OpenAI endpoint is set up correctly.

````python
# Import Libraries
import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI(
    api_key=os.getenv("DEEPSEEK_API_KEY"),
    base_url=os.getenv("DEEPSEEK_BASE_URL")
)

deployment="deepseek-v4-flash"

def get_completion(prompt):
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=deployment,
        messages=messages,
        temperature=0,
        max_tokens=1024
    )

    return response.choices[0].message.content


# Calling the helper function
### 1. Set primary content
text = f"""
What is the current UTC time?
"""

### 2. Use that in the prompt template below
prompt = f"""
```{text}```
"""

### 3. Run the prompt
response = get_completion(prompt)
print(response)
````

### Exercise 5: Complex Prompt
Try a request that has system, user, and assistant messages. System sets assistant context, User and Assistant messages provide multi-turn conversation context.

````python
# Set the completion response based on complex prompt
response = client.chat.completions.create(
    model=deployment,
    messages=[
        {"role": "system", "content": "You are a professional financial analyst."},
        {"role": "user", "content": "What is the market performance of Google in the first quarter of 2026?"},
        {"role": "assistant", "content": "In the first quarter of 2026, Google's parent company Alphabet delivered a strong AI-driven performance, with revenue reaching $109 billion and a net income of $62.6 billion, driving its stock up about 7% in after-hours trading."},
        {"role": "user", "content": "What is the performance of Google compared to other tech-companies in 2026?"}
    ]
)

print(response.choices[0].message.content)
````
